```python
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
```

```python
rating=[]
review_summary=[]
full_review=[]
```

```python
class Mobile_review:
    
    def open_chrome(self):
        self.driver=webdriver.Chrome("C:\web driver\chromedriver.exe")
    
    def open_flipkart(self):
        self.driver.get("{}".format(input("Enter link :")))
        
    def searchs_class_name(self):
        self.driver.find_element_by_class_name("{}".format(input("Enter class name :  "))).send_keys("{}".format(input("Enter item : ")))
        
    def clicks(self):
        self.driver.find_element_by_class_name("{}".format(input("Enter click class "))).click()
        
    def opne_first_iphhone(self):
        self.driver.find_element_by_class_name("{}".format(input("Enter :"))).click()
        
    def opne_all_review(self):
        self.driver.find_element_by_xpath("{}".format(input("Enter class : "))).click()
        
    def rating(self):
        self.rating_class= str(input("Enter rating_class :"))
        for i in self.driver.find_elements_by_xpath("{}".format(self.rating_class)):
            s=i.text
            rating.append(s)
        print(rating)
        
        print(len(rating))
        
    def review(self):
        self.review_class = str(input("Enter class Path of review : "))
        
        for i in self.driver.find_elements_by_xpath("{}".format(self.review_class)):
            r=i.text
            review_summary.append(r)
        print(review_summary)
        print(len(review_summary))
    
    def full_reviews(self):
        self.full_review_class=str(input("Enter class path of full review : "))
        for i in self.driver.find_elements_by_xpath("{}".format(self.full_review_class)):
            r=i.text
            full_review.append(r)
        print(full_review)
        print(len(full_review))
```
