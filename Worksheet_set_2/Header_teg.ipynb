```python
import bs4
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
#Write a python program to display all the header tags from wikipedia.org.
```

```python
rq = requests.get("https://en.wikipedia.org/wiki/Main_Page")
rq
```

```python
scop = BeautifulSoup(rq.content)
scop
```

```python
header =[]
for i in scop.find_all(name='span',class_="mw-headline",recursive=True):
    header.append(i.text)
header
```

```python
import pandas as pd
```

```python
header_df = pd.DataFrame({'Headlines':header})
header_df
```
